```
# Sigmoid Function or Logistic Function Characteristic
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,+5,500)
y = 1 / (1 + np.exp(-x))

plt.plot(40+x, y, color="red")
plt.xlabel('size of tumour')
#plt.color("red")
plt.ylabel('malignant or benign')
plt.show()
```
